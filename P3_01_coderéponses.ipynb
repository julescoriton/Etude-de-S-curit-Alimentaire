```python
import pandas as pd
import numpy as np


data_pop_withA = pd.read_csv("P03/fr_population.csv", sep=",")
filtre_Agrégation_pop = data_pop_withA["Symbole"] != "A"
data_pop = data_pop_withA[filtre_Agrégation_pop]

data_veg_withA = pd.read_csv("P03/fr_vegetaux.csv", sep=",")
filtre_Agrégation_veg = data_veg_withA["Symbole"] != "A"
data_veg = data_veg_withA[filtre_Agrégation_veg]

data_ani_withA = pd.read_csv("P03/fr_animaux.csv", sep=";", encoding="cp1252") #lis fichier encodé cp1252
filtre_Agrégation_ani = data_ani_withA["Symbole"] != "A"
data_ani = data_ani_withA[filtre_Agrégation_ani]



# Question 2 : Identifiez ces redondances, en donnant votre réponse sous forme de formule mathématique
# (pas besoin de coder ici :soleil: ). C'est une équation à 3 termes de type \(a_1 + a2 + [...] = b_1 + b_2 + [...] =
# c_1 + c_2 + [...]\) ) faisant intervenir chacune des 11 quantités données ci dessus. Illustrez cette équation avec 
# l'exemple du blé en France


print("Production + importation + variation de stock - exportation = Disponibilité intérieur = nourriture + semence + pertes + traitement + autre utilisation + aliment pour animaux")


# Question 3 : Calculez (pour chaque pays et chaque produit) la disponibilité alimentaire en kcal puis en kg de protéines.

# 664 Dalim kcal et p
dalim_veg = (data_veg["Code Élément"] == 664) | (data_veg["Code Élément"] == 674) 
dalim_ani = (data_ani["Code Élément"] == 664) | (data_ani["Code Élément"] == 674) 
dalim = (data_veg[dalim_veg]).append(data_ani[dalim_ani]) #rajoute les lignes ani après veg

data_pop_filtrechine = (data_pop["Symbole"] == "A")
data_pop_clean = (data_pop[~data_pop_filtrechine])
pop = (data_pop_clean.drop(columns=["Code Domaine", "Domaine", "Code Élément", "Élément", "Code Produit", "Zone", "Code Produit", "Produit", "Code année", "Année", "Description du Symbole"]))
# suppression de colone avec doublon
dalim_pop = dalim.join(pop.set_index("Code zone"), lsuffix="_Dalim", rsuffix="_pop", on = "Code zone") # rajout info pop en colone



#afficher la valeur ainsi que l'unité calculé
dalim_pop["la disponibilité alimentaire"] =  np.where(dalim_pop['Code Élément']== 664, dalim_pop["Valeur_Dalim"] * (dalim_pop["Valeur_pop"] * 365 / 1000000), 
                                                (np.where(dalim_pop['Code Élément']== 674, dalim_pop["Valeur_Dalim"] * (dalim_pop["Valeur_pop"] * 365 / 1000000), dalim_pop["Valeur_Dalim"] * dalim_pop["Valeur_pop"] * 365 / 1000000)))

dalim_pop['unité'] = np.where(dalim_pop['Code Élément']== 664, "Milliards de Kcal", 
                                                (np.where(dalim_pop['Code Élément']== 674, "Millions de KiloGramme de Proteïne", "")))

print("\n la disponibilité alimentaire en kcal puis en kg de protéines pour chque produit et pays")
dalim_pop

```

```python
# Question 4 : A partir de ces dernières informations, et à partir du poids de la disponibilité alimentaire
# (pour chaque pays et chaque produit), calculez pour chaque produit le ratio "énergie/poids", 
# que vous donnerez en kcal/kg.

# Masque pour sélectionner la nourriture 
masque_5142_veg = (data_veg["Code Élément"] == 5142)
masque_5142_ani = (data_ani["Code Élément"] == 5142)
data_alim_5142 = (data_veg[masque_5142_veg]).append(data_ani[masque_5142_ani]) #fusion de animaux plus vég

# nettoyage de colonne supplémentaire 
data_alim_5142_clean = (data_alim_5142.drop(columns=["Code Domaine", "Domaine", "Code Élément", "Élément", "Zone", "Produit", "Code année", "Année", "Description du Symbole"]))

#rajout de la colonne valeur correspondant à la nourriture en plus de la pop à Dalim
dalim_pop_5142 = pd.merge(dalim_pop, 
             data_alim_5142_clean,
             left_on=["Code Produit","Code zone"],
             right_on=["Code Produit","Code zone"],  how = 'outer')
 
# Calcul du ratio en Kcal/Kg et kg de Proteine/Kg en fonction de la ligne obtenue en multipliant Kcal/p/j par 365 et la population diviser par la nourriture     
dalim_pop_5142["ratio kg dispo"] =  np.where(dalim_pop_5142['Code Élément']== 664, ((dalim_pop_5142["Valeur_Dalim"] * dalim_pop_5142["Valeur_pop"] * 365 * 1000) / (dalim_pop_5142["Valeur"] * 1000000)) , 
                                                                                   ((dalim_pop_5142["Valeur_Dalim"] * dalim_pop_5142["Valeur_pop"] * 365) / (dalim_pop_5142["Valeur"] * 1000000)))



# display units corresponding at each calcul
dalim_pop_5142['unité ratio'] = np.where(dalim_pop_5142['Code Élément']== 664, "Ratio Kcal/Kg", 
                                                (np.where(dalim_pop_5142['Code Élément']== 674, "Ratio Kg de Proteïne/Kg", "")))

dalim_pop_5142_clean = dalim_pop_5142.rename(columns={"Valeur" : "Valeur_nour"}) # rename columns or ligne

print("dataframe du ratio kcal/kg et gramme de proteine /kg \n")
display(dalim_pop_5142_clean)


```
